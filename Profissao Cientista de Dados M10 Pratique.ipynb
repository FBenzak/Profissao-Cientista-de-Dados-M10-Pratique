```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Lendo o arquivo CSV
df = pd.read_csv("ecommerce_preparados.csv")

# Exibir as primeiras linhas
df.head()

# Analise Inicial
# Informações Gerais
df.info()
```

```python
# Estatísticas Descritivas
df.describe()
```

```python
# Verificar Valores Nulos
df.isnull().sum()
```

```python
# Tratando os dados, removendo os valores nulos
df.dropna(inplace=True)

# Verificar Valores Nulos
df.isnull().sum()
```

```python
# GRÁFICO DE DISPERSÃO
plt.figure(figsize=(10,6))

sns.scatterplot(data=df,
                x='Preço',
                y='Nota',
                alpha=0.6)

plt.title('Preço x Nota')
plt.xlabel('Preço')
plt.ylabel('Nota')
plt.show()
```

```python
# MAPA DE CALOR
# Seleciona apenas colunas numéricas

correlacao = df.corr(numeric_only=True)

plt.figure(figsize=(10,6))

sns.heatmap(correlacao,
            annot=True,
            cmap='coolwarm',
            fmt='.2f')

plt.title('Mapa de Calor das Correlações')
plt.show()
```

```python
# GRÁFICO DE BARRAS
# Removendo "Não Definido" da coluna Marca
marcas = df[df['Marca'] != 'Não Definido']['Marca']

plt.figure(figsize=(10,6))

marcas.value_counts().head(10).plot(
    kind='bar',
    color='steelblue'
)

plt.title('Top 10 Marcas com Maior Quantidade de Produtos')
plt.xlabel('Marca')
plt.ylabel('Quantidade de Produtos')
plt.xticks(rotation=45)

plt.show()
```

```python
# GRÁFICO DE PIZZA
# Removendo "Não Definido" da Marca
marcas = df[df['Marca'] != 'Não Definido']['Marca']

# Selecionando as 5 marcas mais frequentes
top_marcas = marcas.value_counts().head(5)

plt.figure(figsize=(8,8))

top_marcas.plot(
    kind='pie',
    autopct='%1.1f%%',
    startangle=90,
    colors=['steelblue', 'orange', 'green', 'red', 'purple']
)

plt.title('Distribuição das 5 Principais Marcas do Catálogo')
plt.ylabel('')

plt.show()
```

```python
# Removendo valores nulos da coluna Preço
precos = df[df['Preço'].notna()]['Preço']

plt.figure(figsize=(10,6))

sns.kdeplot(
    x=precos,
    fill=True,
    color='green'
)

plt.title('Densidade dos Preços dos Produtos')
plt.xlabel('Preço')
plt.ylabel('Densidade')

plt.show()
```
